# Demonstration of the survey module

This tests a quick and dirty refactorizaton of the functionality in the FRF_2012_ProfileSurvey_DATA/ReadData.ipynb  notebook into the survey module. I also got rid of the line-based parsing because I'm lazy.

First set up the notebook to do inline plotting and import our survey module 

```python
%matplotlib inline
import survey
```

Now instantiate a SurveySite object by telling it to look in a data directory

```python
ss = survey.SurveySite("FRF_2012_ProfileSurvey_Data")
```

The surveys data member contains the list of Survey objects (instances) that were constructed  from the csv files in the directory

```python
ss.surveys
```

Now check if the member functions of the Survey instances  work (just testing on one survey)

```python
plt = ss.surveys[0].plot_lines()
```

The lines data member is a dictionary of the survey lines of a particular survey

```python
ss.surveys[0].lines.keys()
```

You can plot any subset of the lines by passing in a list

```python
plt = ss.surveys[0].plot_lines([135,163,190])
```

```python
plt = ss.surveys[0].plot_contours()
```

```python
plt = ss.surveys[0].plot_surface()
plt.show()
```

```python
ss.surveys[-1].kml("FRF")
```

You can generate an animation of the contour plots of the survey by saving each frame

```python
ss.contourAnimation()
```
